# Scraping Notebook

## A. Import Libraries

```python
# import package
from selenium import webdriver
from bs4 import BeautifulSoup


import pandas as pd
import numpy as np

import time
```

## B. Scraping Process

### 1. Scraping Somethinc Product

```python
# variable
name_list = []
price_list_discount = []
discount_list = []
merchant_list = []
city_list = []
sales_list = []
rating_list = []


driver = webdriver.Chrome()

for num in range(1, 51):
    # url link
    url = f'https://www.tokopedia.com/search?navsource=&page={num}&q=somethinc&srp_component_id=02.01.00.00&srp_page_id=&srp_page_title=&st='

    # access website
    driver.get(url)

    # spare time loading
    time.sleep(10)

    # fetch html
    html = driver.page_source

    # parse
    soup = BeautifulSoup(html, "html.parser")

    boxes = soup.find_all('div', {'class' : 'gG1uA844gIiB2+C3QWiaKA=='})

    for box in boxes:
        
        # scrapping product_name
        try:
            name_list.append(box.find('span', {'class' : '+tnoqZhn89+NHUA43BpiJg=='}).get_text())
        except:
            name_list.append(None)

        # scrapping product price discount
        try:
            price_list_discount.append(box.find('div', {'class' : 'urMOIDHH7I0Iy1Dv2oFaNw== HJhoi0tEIlowsgSNDNWVXg=='}).get_text())
        except:
            price_list_discount.append(None)

        # scraping discount percentage
        try:
            discount_list.append(box.find('span', {'class': '_7UCYdN8MrOTwg0MKcGu8zg=='}).get_text())
        except:
            discount_list.append(None)

        # scraping merchant name
        try:
            merchant_list.append(box.find('span', {'class' : 'si3CNdiG8AR0EaXvf6bFbQ== gxi+fsEljOjqhjSKqjE+sw== flip'}).get_text())
        except:
            merchant_list.append(None)

        # scraping city name
        try:
            city_list.append(box.find('span', {'class' : 'gxi+fsEljOjqhjSKqjE+sw== flip'}).get_text())
        except:
            city_list.append(None)

        # scraping unit sold
        try:
            sales_list.append(box.find('span', {'class' : 'u6SfjDD2WiBlNW7zHmzRhQ=='}).get_text())
        except:
            sales_list.append(None)

        # scraping product rating
        try:
            rating_list.append(box.find('span', {'class' : '_2NfJxPu4JC-55aCJ8bEsyw=='}).get_text())
        except:
            rating_list.append(None)


# create dataframe from list
somethinc_scrap = {'product_name' : name_list,
        'price_after_discount': price_list_discount,
        'discount_percentage': discount_list,
        'merchant' : merchant_list,
        'merchant_city' : city_list,
        'unit_sold' : sales_list,
        'product_rating' : rating_list}

# store in raw_data variable
somethinc_raw = pd.DataFrame(somethinc_scrap)

# drop unfillable data
somethinc_raw = somethinc_raw.dropna(subset=['product_name', 'price_after_discount', 'unit_sold'])

# save data to csv
somethinc_raw.to_csv('somethinc-raw.csv', index=False)
```
